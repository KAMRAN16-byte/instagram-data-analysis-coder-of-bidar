### Load data store in .txt extension file.


```python
with open('raw data.txt', encoding='utf-8') as f:
  data = f.read()
```

### Processioning Data into small Chunks.

```python
def parse_chunk(chunk):
  chunk = chunk.strip()
  if len(chunk) > 5:
    try:
      sub_chunk = chunk.split('\n')
      if len(sub_chunk) < 6:
        sub_chunk.append('No bio of this user')

    #For Posts.
      sub_chunk[1] = sub_chunk[1].replace(',','')
      if 'K' in sub_chunk[1]:
        sub_chunk[1] = int(float(sub_chunk[1].split(' ')[0].replace('K','')) * 1000)
      elif 'M' in sub_chunk[1]:
        sub_chunk[1] = int(float(sub_chunk[1].split(' ')[0].replace('M','')) * 1000000)
      else:
        sub_chunk[1] = int(sub_chunk[1].split(' ')[0])

    #For Followers.
      sub_chunk[2] = sub_chunk[2].replace(',','')
      if 'K' in sub_chunk[2]:
        sub_chunk[2] = int(float(sub_chunk[2].split(' ')[0].replace('K','')) * 1000)
      elif 'M' in sub_chunk[2]:
        sub_chunk[2] = int(float(sub_chunk[2].split(' ')[0].replace('M','')) * 1000000)
      else:
        sub_chunk[2] = int(sub_chunk[2].split(' ')[0])

    #For Following.
      sub_chunk[3] = sub_chunk[3].replace(',','')
      if 'K' in sub_chunk[3]:
        sub_chunk[3] = int(float(sub_chunk[3].split(' ')[0].replace('K','')) * 1000)
      elif 'M' in sub_chunk[3]:
        sub_chunk[3] = int(float(sub_chunk[3].split(' ')[0].replace('M','')) * 1000000)
      else:
        sub_chunk[3] = int(sub_chunk[3].split(' ')[0])

      #Return in dictionary.
      return {"username":sub_chunk[0], "no_of_posts":sub_chunk[1], "no_of_follower":sub_chunk[2],"no_of_following":sub_chunk[3], "name":sub_chunk[4], "Bio":'\n'.join(sub_chunk[5:])}
    except Exception as e:
      print(e)
      return None
  return None
```

### Analysis the preprocessed data.

```python
max_follower = 0
max_post = 0
max_following = 0
for dataset in data:
  if(max_post<dataset['no_of_posts']):
    max_post = dataset['no_of_posts']
    data_max_posts = dataset
  if(max_follower<dataset['no_of_follower']):
    max_follower = dataset['no_of_follower']
    data_max_follower = dataset
  if(max_following<dataset['no_of_following']):
    max_following = dataset['no_of_following']
    data_max_following = dataset
```
